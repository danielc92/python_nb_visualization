# Seaborn tests

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import io
import numpy as np
%matplotlib inline

sns.set(rc={'figure.figsize':(12,9)})
sns.set_style('whitegrid')
sns.set_palette('Greys')
```

```python
#1. Prepare data
#2. Control figure aesthetics
#3. Plot with seaborn
#4. Further customize plot
```

## Categorical plots with Seaborn

#### Scatterplot

```python
sns.stripplot(data = iris,
             x = "species",
             y = "petal_length")
```

#### Swarmplot

```python
sns.swarmplot(data = iris,
             x = "species",
             y = "petal_length")
```

#### Bar Chart

```python
sns.barplot(data = titanic, 
           x = "sex",
           y = "survived",
           hue = "class")
```

#### Count Plot

```python
sns.countplot(data = titanic,
             x = "deck")

plt.ylabel("COUNT OF RECORDS")
plt.xlabel("TYPE OF HUMAN")
plt.title("PASSENGER TYPES")
```

#### Point Plot

```python
sns.pointplot(data = titanic,
             x = "class",
             y = "survived",
             hue = "sex",
             pallette = {'male':'g', 'female':'m'},
             markers = ["^", "o"],
             linestyles = ["-", "--"])
```

#### Boxplot

```python
sns.boxplot(data = titanic,
           x = 'alive',
           y = 'age',
           hue = 'pclass',
           )
```

#### Violinplot

```python
sns.violinplot(data = titanic,
              x = 'age',
              y = 'sex',
              hue = "survived")
```

## Regression Plots

```python
sns.regplot(data = titanic,
           y = 'fare',
           x = 'age')
```

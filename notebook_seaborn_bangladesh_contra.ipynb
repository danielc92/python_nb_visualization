```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]

seaborn.set_style("white")
seaborn.set_palette(flatui)
```

### Import Data 

```python
data = pandas.read_csv("/users/danielcorcoran/Desktop/bangl.csv")
```

```python
data.head()
```

```python
data = pandas.concat([data, pandas.cut(data["age"],bins = 8).rename("age_bins")], axis = 1)
```

```python
seaborn.countplot(data = data, x = "age_bins")\
.get_figure().savefig("/users/danielcorcoran/Desktop/github_repos/python_nb_visualization/seaborn_exports/bangladesh2.svg")
```

```python
seaborn.factorplot(x = "use_contracep",
                   data = data, 
                   col_wrap = 2,
                   col = "living_children", 
                   kind = "count", 
                   size = 6.4)\
.savefig("/users/danielcorcoran/Desktop/github_repos/python_nb_visualization/seaborn_exports/bangladesh1.svg")
```

```python
seaborn.factorplot(x = "age_bins",
                   data = data, 
                   col_wrap = 2,
                   col = "living_children", 
                   kind = "count", 
                   size = 6.5)\
.savefig("/users/danielcorcoran/Desktop/github_repos/python_nb_visualization/seaborn_exports/bangladesh3.svg")
```

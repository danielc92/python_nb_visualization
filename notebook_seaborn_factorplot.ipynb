## Examples

### Factorplot bar using tips dataset

```python
seaborn.factorplot(data = tips,
                  x = "day",
                  y = "total_bill",
                  kind = "bar",
                  col = "sex",
                  legend = True)
```

### Factorplot point using exercise dataset

```python
seaborn.factorplot(data = exercise, 
                  x = "time",
                  y = "pulse",
                  #hue = "kind", 
                  col = "diet",
                  legend = True,
                  estimator = sum)
```

### Factorplot point using exercise dataset, splitting data by 'kind'

```python
seaborn.factorplot(data = exercise, 
                  x = "time",
                  y = "pulse",
                  hue = "kind", 
                  col = "diet",
                  legend = True)
```

### Factorplot bar using titanic dataset
- 'bar' kind will average by default

```python
seaborn.factorplot(data = titanic,
                  x = "sex",
                  y = "survived",
                  kind = "bar",
                  #estimator = sum,
                  legend = True)
```

### Adding estimator parameter (sum) to Factorplot above gives the sum of 'survived'

```python
seaborn.factorplot(data = titanic,
                  x = "sex",
                  y = "survived",
                  kind = "bar",
                  estimator = sum,
                  legend = True)
```

### Factorplot count of titanic dataset splitting by surived

```python
seaborn.factorplot(data = titanic,
                  x = "sex",
                  col = "survived",
                  kind = "count",
                  estimator = sum,
                  legend = True)
```

### Factorplot count of titanic dataset splitting by surived

```python
seaborn.factorplot(data = titanic,
                  x = "class",
                  y = "survived",
                  kind = "bar",
                  estimator = sum,
                  col = "sex")
```

### Factorplot bar of tips dataset, splitting by 'time', with custom palette

```python
seaborn.factorplot(data = tips,
                  x = "sex",
                   y = "total_bill",
                   kind = "bar",
                  col = "time",
                  palette = ["orange", "purple"])
```

### Factorplot swarm of tips dataset, splitting by 'day' 

```python
seaborn.factorplot(data = tips,
                  x = "sex",
                  y = "tip",
                  kind = "swarm",
                   hue = "smoker",
                  col = "day",
                  col_wrap = 2)
```

### Appending column to tips dataset to bin tips

```python
tips["tip_bracket"] = pandas.cut(x = tips["tip"],bins=5)
```

### Factorplot strip (scatter) using tips dataset, splitting by 'day'

```python
seaborn.factorplot(data = tips,
                 x = 'tip_bracket',
                 y = 'total_bill',
                 hue = "sex",
                 kind = "strip",
                 col = "day", 
                 col_wrap = 2)
```

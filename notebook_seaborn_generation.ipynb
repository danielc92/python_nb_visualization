```python
measures = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
categories = ['species']
```

## Process

```python
x = "species"

for measure in measures:
    
    plt.figure()
    
    print("processing measure:",measure)
    
    title = "Average {} by {} \n using iris dataset".format(x, measure)
    
    chart = seaborn.barplot(data = data,
                       x = x,
                       y = measure,
                       palette = "Set2")\
                        .set_title(title)
    
    plt.savefig(export_path + "figure_{}_{}.svg".format(x,measure))
```

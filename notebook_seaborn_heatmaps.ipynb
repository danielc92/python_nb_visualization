## Set figure size global

```python
seaborn.set(rc={'figure.figsize':(12,9)})
seaborn.set_style('whitegrid')
```

## Create random data for testing heatmaps

```python
random_data = pandas.DataFrame(numpy.random.rand(12,12))
```

```python
random_data
```

## Create rounded down version of test data

```python
random_data_rounded = random_data.round(1)
```

## Examples

### Heatmap basic with custom cmap

```python
seaborn.heatmap(data = random_data, 
                cmap = "summer")
```

### Heatmap with legend disabled

```python
seaborn.heatmap(data = random_data, 
                cmap = "summer",
               cbar = False)
```

### Heatmap with legend truncation

```python
seaborn.heatmap(data = random_data, 
                cmap = "bwr",
               vmin = .4,
               vmax = .8)
```

### Heatmap with value labels, using rounded test data

```python
seaborn.heatmap(data = random_data_rounded, 
                cmap = "PuRd",
               annot = True)
```

### Heatmap with linewidth of 0.8

```python
seaborn.heatmap(data = random_data_rounded, 
                cmap = "Accent_r",
               linewidth = 0.8)
```

### Heatmap using correlation coefficients from test data, including save to disk
- Note: export as .svg for great scalability 

```python
corr = random_data.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

seaborn.heatmap(corr, 
                cmap = "jet_r",
                mask=mask, 
                square=True).get_figure().savefig("heatmap.svg")

```
